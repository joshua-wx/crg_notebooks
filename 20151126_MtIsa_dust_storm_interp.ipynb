```python
#imports
from __future__ import print_function
import pyart
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from time import time
%matplotlib inline
```

```python
fn  = 'data/201511261606Mnt_Isa.h5'
#Since the names in the HDF5 file are non-standard we use file_field_names = True
radar = pyart.aux_io.read_odim_h5(fn, file_field_names=True) 
```

```python
t1 = time()
grids  = pyart.map.grid_from_radars(
        radar, grid_shape=(61, 401, 401),
        grid_limits=((0, 10000.0),(-100000, 100000), (-100000, 100000)),
        fields=['DBZH'], gridding_algo="map_gates_to_grid")
print(time() - t1)
```

```python
display = pyart.graph.GridMapDisplay(grids)
fig = plt.figure(figsize=[15, 7])

# panel sizes
map_panel_axes = [0.0, 0.0, 0.6, 1]
x_cut_panel_axes = [0.6, 0.0, 0.4, 0.4]
y_cut_panel_axes = [0.6, 0.6, 0.4, 0.4]

#lat/lon lines
lat_lines = np.arange(-21.5, -19.5, 0.5)
lon_lines = np.arange(138, 141, 0.5)

# parameters
level = 1
vmin = -8
vmax = 64
lon  = 139.5552
lat  = -20.7112

# panel 1
ax1 = fig.add_axes(map_panel_axes)
display.plot_basemap(lat_lines = lat_lines, lon_lines = lon_lines)
display.plot_grid('DBZH', level=level, vmin=vmin, vmax=vmax,
                 cmap = pyart.graph.cm.NWSRef)
display.plot_crosshairs(lon=lon, lat=lat)

# panel 2, longitude slice.
ax2 = fig.add_axes(x_cut_panel_axes)
ax2.set_ylim([0,5])
ax2.set_xlim([-100,100])
display.plot_longitude_slice('DBZH', lon=lon, lat=lat, vmin=vmin, vmax=vmax,
                            cmap = pyart.graph.cm.NWSRef)
# panel 3, latitude slice
ax3 = fig.add_axes(y_cut_panel_axes)
ax3.set_ylim([0,5])
ax3.set_xlim([-100,100])
display.plot_latitude_slice('DBZH', lon=lon, lat=lat, vmin=vmin, vmax=vmax,
                           cmap = pyart.graph.cm.NWSRef)

```
