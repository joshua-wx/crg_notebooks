```python
#imports
from __future__ import print_function
import pyart
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from time import time
%matplotlib inline
```

```python
fn  = 'data/201511261606Mnt_Isa.h5'
#Since the names in the HDF5 file are non-standard we use file_field_names = True
radar = pyart.aux_io.read_odim_h5(fn, file_field_names=True) 
```

```python
xsect = pyart.util.cross_section_ppi(radar, [0, 30])

display = pyart.graph.RadarDisplay(xsect)
fig = plt.figure(figsize=[20,20])
ax = fig.add_subplot(211)
display.plot_rhi('DBZH', 0, vmin=0, vmax=65)
display.set_limits(ylim=[0, 10], xlim=(0, 100))
ax = fig.add_subplot(212)
display.plot_rhi('DBZH', 1, vmin=0, vmax=65)
display.set_limits(ylim=[0, 10], xlim=(0, 100))
plt.tight_layout()
plt.show()
```
