Load required libraries

```python
#imports
from __future__ import print_function
import pyart
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import numpy as np
import os

```

### Plot single SPPI

Load odimh5 file from directory into myradar object (modifiy path to file as needed)

```python
fn  = '/home/meso/temp/h5_20151210/uq-xpol_sppi_20151210_163235.h5'
#Since the names in the HDF5 file are non-standard we use file_field_names = True
myradar = pyart.aux_io.read_odim_h5(fn, file_field_names=True) 
```

create grid lines and correct myradar azimuths to sector PPI (fix required)

```python
lat_lines = np.arange(-27.5,-28.5,-.1)
lon_lines = np.arange(152.5, 153.5, .1)
my_display = pyart.graph.RadarMapDisplay(myradar)
my_display.azimuths=np.linspace(120,270,346)
```

Plot DBZH and VRADH wiht a marker for Beaudesert

```python
font = {'size': 18}
matplotlib.rc('font', **font)
f = plt.figure(figsize=[18,18])
plt.subplot(2,1,1)
my_display.plot_ppi_map('DBZH', vmin=0, vmax=50, 
                        lat_lines = lat_lines, lon_lines = lon_lines,
                        max_lat = -27.95, min_lat =-28.1, min_lon = 152.8, max_lon = 153.1,
                       cmap = pyart.graph.cm.NWSRef)
my_display.plot_point( 152.983333, -27.983333,
               label_text = 'Beaudesert', 
               label_offset=[.001,.001], symbol = 'wo')

plt.subplot(2,1,2)
my_display.plot_ppi_map('VRADH', vmin=-25, vmax=25, 
                        lat_lines = lat_lines, lon_lines = lon_lines,
                        max_lat = -27.95, min_lat =-28.1, min_lon = 152.8, max_lon = 153.1,
                       cmap = pyart.graph.cm.EWilson17)
my_display.plot_point( 152.983333, -27.983333,
               label_text = 'Beaudesert', 
               label_offset=[.001,.001], symbol = 'wo' )
```

### Process a sequence of radar files

load directory listing (change as needed to dir containing odimh5 files)

```python
dirr = '/home/meso/temp/h5_20151210/'
fls = os.listdir(dirr)
fls.sort()
```

loop through file list and export plots to file

```python

i = 0
for fl in fls:
    try:
        print('doing ', dirr+ fl)
        myradar = pyart.aux_io.read_odim_h5(dirr+ fl, file_field_names=True)
        
        lat_lines = np.arange(-27.5,-28.5,-.1)
        lon_lines = np.arange(152.5, 153.5, .1)
        my_display = pyart.graph.RadarMapDisplay(myradar)
        my_display.azimuths=np.linspace(120,270,346)
        
        font = {'size': 18}
        matplotlib.rc('font', **font)
        f = plt.figure(figsize=[18,18])
        plt.subplot(2,1,1)
        my_display.plot_ppi_map('DBZH', vmin=0, vmax=50, 
                                lat_lines = lat_lines, lon_lines = lon_lines,
                                max_lat = -27.95, min_lat =-28.1, min_lon = 152.8, max_lon = 153.1,
                                cmap = pyart.graph.cm.NWSRef, resolution = 'h')
        
        my_display.plot_point( 152.983333, -27.983333,
               label_text = 'Beaudesert', 
               label_offset=[.001,.001], symbol = 'wo' )
        
        ax = plt.gca()
        plt.subplot(2,1,2)
        my_display.plot_ppi_map('VRADH', vmin=-25., vmax=25., 
                                lat_lines = lat_lines, lon_lines = lon_lines,
                                max_lat = -27.95, min_lat =-28.1, min_lon = 152.8, max_lon = 153.1,
                                cmap = pyart.graph.cm.EWilson17, colorbar_label = 'Radial Velocity (m/s)',
                                resolution = 'h')
        
        my_display.plot_point( 152.983333, -27.983333,
               label_text = 'Beaudesert', 
               label_offset=[.001,.001], symbol = 'wo' )
        
        ax = plt.gca()
        plt.savefig('/home/meso/temp/images_20151210/ftorn_%03.0f.png' %i, dpi=100)
        plt.close()
        i = i +1
    except:
        pass
```
