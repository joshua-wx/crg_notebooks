```python
#imports
from __future__ import print_function
import pyart
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import numpy as np
from time import time
import os
from netCDF4 import num2date, date2num
```

```python
lat_lines = np.arange(-27.5,-28.5,-.1)
lon_lines = np.arange(152.0, 153.5, .1)
```

## 17-12-2014 (tx pulse 1) 0.5deg

```python
fn  = '/home/meso/temp/h5/uq-xpol_scn_00_01_20141217_155200.h5'
myradar = pyart.aux_io.read_odim_h5(fn, file_field_names=True) 
my_display = pyart.graph.RadarMapDisplay(myradar)
my_display.azimuths = np.linspace(0,360,2000)
f = plt.figure(figsize=[20,18])
my_display.plot_ppi_map('DBZH', vmin=0, vmax=50, 
                        lat_lines = lat_lines, lon_lines = lon_lines,
                        max_lat = -27.5, min_lat =-28.5, min_lon = 152.0, max_lon = 153.0,
                       cmap = pyart.graph.cm.NWSRef)
```
