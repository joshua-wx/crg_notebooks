```python
#imports
from __future__ import print_function
import pyart
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from time import time
import os
%matplotlib inline
```

```python
def plot_panel(in_fn,out_fn,field):

    #Since the names in the HDF5 file are non-standard we use file_field_names = True
    radar = pyart.aux_io.read_odim_h5(in_fn, file_field_names=True) 
    #correct field names
    radar.fields['DBZH']['standard_name'] = 'Reflectivity'
    radar.fields['DBZH']['units'] = 'dBZ'
    radar.fields['DBZH']['long_name'] = 'Radar Reflectivity Factor'
    radar.fields['VRADH']['standard_name'] = 'Velocity'
    radar.fields['VRADH']['units'] = 'm/s'
    radar.fields['VRADH']['long_name'] = 'Radial Velocity of Scatterers'

    #create figure
    fig = plt.figure(figsize=[15, 7])

    # panel sizes
    map_panel_axes = [0.05, 0.05, 0.5, 0.9]
    rhi1_panel_axes = [0.6, 0.575, 0.35, 0.35]
    rhi2_panel_axes = [0.6, 0.1, 0.35, 0.35]


    #lat/lon lines
    lat_lines = np.arange(-21.5, -19.5, 0.5)
    lon_lines = np.arange(138, 141, 0.5)

    # parameters
    level = 1
    if field=='DBZH':
        vmin = -8
        vmax = 64
        cmap = pyart.graph.cm.NWSRef

    if field=='VRADH':
        vmin = -15
        vmax = 15
        cmap = pyart.graph.cm.BuDRd18  

    max_lat  = -19.81
    min_lat  = -21.61
    min_lon  = 138.59
    max_lon  = 140.52
    r_lat    = -20.7112
    r_lon    = 139.5552
    horz_rng = 100
    vert_rng = 5
    tilt = 0
    rhi1_az = 0 #note, these are all inverted
    rhi2_az = 180

    #create objects
    xsect = pyart.util.cross_section_ppi(radar, [rhi1_az, rhi2_az])
    rhi_display = pyart.graph.RadarDisplay(xsect)
    map_display = pyart.graph.RadarMapDisplay(radar)

    # panel 1
    ax1 = fig.add_axes(map_panel_axes)
    map_display.plot_ppi_map(field, tilt, vmin=vmin, vmax=vmax, 
                            lat_lines = lat_lines, lon_lines = lon_lines,
                            max_lat = max_lat, min_lat = min_lat, min_lon = min_lon, max_lon = max_lon,
                            cmap = cmap)
    map_display.plot_range_rings([1,50,100], ax=ax1, col='k', ls='-', lw=0.5)
    map_display.plot_point( r_lon, r_lat, symbol = 'wo' )
    map_display.plot_line_geo([r_lon,r_lon], [-20.27,-21.17], line_style='k--', lw=3)

    # panel 2, slice.
    ax2 = fig.add_axes(rhi1_panel_axes)
    rhi_display.plot_rhi(field, 0, vmin=vmin, vmax=vmax, title='0 Deg. RHI',cmap = cmap)
    rhi_display.set_limits(ylim=[0, vert_rng], xlim=(0, 50))
    rhi_display.plot_grid_lines(ax=ax2, col='k', ls=':')

    # panel 2, slice.
    ax3 = fig.add_axes(rhi2_panel_axes)
    rhi_display.plot_rhi(field, 1, vmin=vmin, vmax=vmax, title='180 Deg. RHI',cmap = cmap)
    rhi_display.set_limits(ylim=[0, vert_rng], xlim=(0, 50))
    rhi_display.plot_grid_lines(ax=ax3, col='k', ls=':')

    # save figure
    plt.savefig(out_fn, dpi=100)
```

```python
plot_panel('data/201511261336Mnt_Isa.h5','tmp/MtIsaDemo.png','DBZH')
```
