```python
#imports
from __future__ import print_function
import pyart
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import numpy as np
import os
from netCDF4 import num2date, date2num
import cartopy.crs as ccrs
```

### Plot 2016 dataset

```python
fn    = '/home/meso/Desktop/testing-middlerd_201710/20161221-odimh5/99_20161221_061200.h5'
radar = pyart.aux_io.read_odim_h5(fn, file_field_names=True) 
display = pyart.graph.RadarMapDisplayCartopy(radar)

lat_lines = np.arange(-26,-28,-.1)
lon_lines = np.arange(152, 154, .1)

font = {'size': 14}
matplotlib.rc('font', **font)

f = plt.figure(figsize=[20,40])
ax = plt.subplot(411, projection=ccrs.PlateCarree())
display.plot_ppi_map('DBZH', vmin=0, vmax=70, ax=ax, 
                        lat_lines = lat_lines, lon_lines = lon_lines,
                        cmap = pyart.graph.cm.NWSRef,resolution = '10m')

ax = plt.subplot(412, projection=ccrs.PlateCarree())
display.plot_ppi_map('VRADH', vmin=-15, vmax=15, ax=ax,
                        lat_lines = lat_lines, lon_lines = lon_lines,
                       cmap = 'pyart_BuDRd18',resolution = '10m')

ax = plt.subplot(413, projection=ccrs.PlateCarree())
display.plot_ppi_map('ZDR', vmin=-1, vmax=6, ax=ax,
                        lat_lines = lat_lines, lon_lines = lon_lines,
                       cmap = 'jet',resolution = '10m')


ax = plt.subplot(414, projection=ccrs.PlateCarree())
display.plot_ppi_map('KDP', vmin=-1, vmax=6, ax=ax,
                        lat_lines = lat_lines, lon_lines = lon_lines,
                       cmap = 'jet',resolution = '10m')
```
